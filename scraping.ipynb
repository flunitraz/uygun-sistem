```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

gaming.gen.tr

```python
ggt_pages = ["https://www.gaming.gen.tr/kategori/hazir-sistemler/page/1/",
             "https://www.gaming.gen.tr/kategori/hazir-sistemler/page/2/",
             "https://www.gaming.gen.tr/kategori/hazir-sistemler/page/3/"]
ggt_data=[]
```

```python
for ggt_page in ggt_pages:   
    response = requests.get(ggt_page)
    soup = bs(response.text, "html.parser")
    d = soup.find("ul","products columns-3").find_all("a")
    for url in d:
        ggt_data.append(url.get("href"))
```

```python
ggt_data_list = []

for d in ggt_data:
    ggt_dict = {}
    response = requests.get(d)
    soup = bs(response.text, "html.parser")
    if(soup.find("p","price").text!=""):
        for i in range(7):
            ggt_dict[soup.find_all("td")[i*3].text] = soup.find_all("td")[(i*3)+1].text

        ggt_dict["Fiyat"] = soup.find("p","price").text.split()[-2]
        ggt_dict["Satıcı"] = "Gaming Gen TR"
        ggt_dict["Url"] = d
    ggt_data_list.append(ggt_dict)
```

```python
df_gaming_gen_tr = pd.DataFrame(ggt_data_list)
df_gaming_gen_tr = df_gaming_gen_tr[["Anakart","İşlemci","Ekran Kartı","RAM Bellek","SSD","Kasa","Fiyat","Satıcı","Url"]].dropna()
```

itopya.com

```python
itopya_pages = ["https://itopya.com/HazirSistemler/?sayfa=1&stok=stokvar",
           "https://itopya.com/HazirSistemler/?sayfa=2&stok=stokvar"]
itopya_data = []
itopya_prices = []
```

```python
for i_page in itopya_pages:   
    response = requests.get(i_page)
    soup = bs(response.text, "html.parser")
    d = soup.find_all("div","product-header")
    for ind,url in enumerate(d):
        itopya_data.append("https://itopya.com"+url.find("a").get("href"))
        itopya_prices.append(soup.find_all("div","price")[ind].text.strip())  
```

```python
itopya_data_list = []

for ind,d in enumerate(itopya_data):
    itopya_dict = {}
    response = requests.get(d)
    soup = bs(response.text, "html.parser")
    
    if(len(soup.find_all("div","product-card active","name secLink"))==7):
        for i in range(6):
            itopya_dict[i] = soup.find_all("div","product-card active","name secLink")[i+1].text.replace("\n","").replace("Seçildi","").replace("Fark:+0,00","").replace("\r","").replace("-+","").strip()

        itopya_dict["Fiyat"] = itopya_prices[ind]
        itopya_dict["Satıcı"] = "İtopya"
        itopya_dict["Url"] = d
    itopya_data_list.append(itopya_dict)
```

```python
df_itopya = pd.DataFrame(itopya_data_list)
df_itopya = df_itopya[[1,0,3,2,5,4,"Fiyat","Satıcı","Url"]].dropna()
```

```python
df_gaming_gen_tr.columns=["Anakart","İşlemci","Ekran Kartı","RAM Bellek","SSD","Kasa","Fiyat","Satıcı","Url"]
df_itopya.columns=["Anakart","İşlemci","Ekran Kartı","RAM Bellek","SSD","Kasa","Fiyat","Satıcı","Url"]
```
